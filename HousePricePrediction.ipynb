```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv('E:/housing.csv')
```

```python
df.dropna(inplace=True)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop(['median_house_value'],axis=1)
y = df['median_house_value']
```

```python
X
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)
```

```python
train_data = X_train.join(y_train)
```

```python
train_data
```

```python
train_data.drop(['ocean_proximity'],axis=1).corr()
```

```python
train_data
```

```python
sns.heatmap(train_data.drop(['ocean_proximity'],axis=1).corr(),annot=True,cmap='YlGnBu')
```

```python
# We have applied the log on the rightly skewed data, to make the data much more symetrical.
# if the data is left skewed we can use squaring or cubing
train_data['total_rooms'] = np.log(train_data['total_rooms'] + 1)
train_data['total_bedrooms'] = np.log(train_data['total_bedrooms'] + 1)
train_data['population'] = np.log(train_data['population'] + 1)
train_data['households'] = np.log(train_data['households'] + 1)
```

```python
train_data.hist(figsize=(15,8))
plt.show()
```

```python
train_data['ocean_proximity'].value_counts()
```

```python
# converting the categorical column to one-hot encoded column
train_data = train_data.join(pd.get_dummies(train_data['ocean_proximity']).astype(int)).drop(['ocean_proximity'],axis=1)
```

```python
train_data
```

```python
plt.figure(figsize=(20,10))
sns.heatmap(train_data.corr(),annot=True)
```

```python
plt.figure(figsize=(15,8))
sns.scatterplot(x='latitude',y='longitude',data=train_data,hue='median_house_value',palette='coolwarm')
```

```python
train_data['bedroom_ratio'] = train_data['total_bedrooms'] / train_data['total_rooms']
```

```python
train_data['household_rooms'] = train_data['total_rooms']/train_data['households']
```

```python
plt.figure(figsize=(20,10))
sns.heatmap(train_data.corr(),annot=True)
```

```python
from sklearn.linear_model import LinearRegression
X_train,y_train = train_data.drop(['median_house_value'],axis=1), train_data['median_house_value']
```

```python
reg = LinearRegression()
```

```python
reg.fit(X_train, y_train)
```

```python
test_data = X_test.join(y_test)

test_data['total_rooms'] = np.log(test_data['total_rooms'] + 1)
test_data['total_bedrooms'] = np.log(test_data['total_bedrooms'] + 1)
test_data['population'] = np.log(test_data['population'] + 1)
test_data['households'] = np.log(test_data['households'] + 1)

test_data = test_data.join(pd.get_dummies(test_data['ocean_proximity']).astype(int)).drop(['ocean_proximity'],axis=1)

test_data['bedroom_ratio'] = test_data['total_bedrooms'] / test_data['total_rooms']
test_data['household_rooms'] = test_data['total_rooms'] / test_data['households']
```

```python
X_test,y_test = test_data.drop(['median_house_value'],axis=1), test_data['median_house_value']
```

```python
reg.score(X_test, y_test)
```

```python
from sklearn.ensemble import RandomForestRegressor


forest = RandomForestRegressor()

forest.fit(X_train, y_train)
```

```python
forest.score(X_test, y_test)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
forest = RandomForestRegressor()
```

```python
param_grid = {
    "n_estimators" : [100, 200, 300],
    "min_samples_split" : [2, 4],
    "max_depth" : [None, 4, 8]
}
grid_search = GridSearchCV(forest, param_grid, cv=5,
                           scoring ='neg_mean_squared_error',
                           return_train_score=True
    
)
grid_search.fit(X_train,y_train)
```

```python
model = grid_search.best_estimator_
```

```python
model.score(X_test,y_test)
```

```python
import pickle
```

```python
pickle.dump(model,open('model.pkl','wb'))
```

```python
load_model = pickle.load(open('model.pkl','rb'))
```
